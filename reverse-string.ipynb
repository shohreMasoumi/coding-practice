## Pay Attention To The Details

__Q__: Would the input string contain white spaces? <br>
__A__: Yes. It is possible.

__Q__: In the venet of leading and following white spaces, how should the algorithm perform? <br>
__A__: There should be no leading or following white spaces in the reversed string

````python
def reverse_string(s):
    N = len(s)
    i, j = 0, N - 1
    while s[i] == ' ':
        i += 1
    while s[j] == ' ':
        j -= 1
    s = s[i: (j + 1)]
    i, j = i - i, j - i
    while i < j:
        temp = s[i]
        s[i] = s[j]
        s[j] = temp
        i += 1
        j -= 1
    
    return s

print(reverse_string(list("Hello")))
print(reverse_string(list("  Hello")))
print(reverse_string(list("  Hellow     ")))

# Time scomplexity: O(N)
# Space complexity: O(1)
````

# Reverse The Order of the Words

````python
def reverse_words_orders(s):
    N = len(s)
    i, j = 0, N - 1
    while s[i] == ' ':
        i += 1
    while s[j] == ' ':
        j -= 1
    s = s[i:(j + 1)] 
    i, j = i - i, j - i
    N = len(s)

    s += [' ']
    N += 1
    start, end, white_space_loc = 0, 0 ,0
    while white_space_loc <  N:
        if s[white_space_loc] == ' ':
            end = white_space_loc - 1
            while start < end:
                temp = s[start]
                s[start] = s[end]
                s[end] = temp
                start += 1
                end -= 1
            start = white_space_loc + 1
        white_space_loc += 1
    
    s = s[:(N - 1)]
    return s[::-1]

print(''.join(reverse_words_orders(list("   Hello World  "))))
print()
print(''.join(reverse_words_orders(list("   Hello World!?  "))))
print()
print(''.join(reverse_words_orders(list("   Hello   World!?  "))))
print()

# Time complexity: O(N)
# Space complexity: O(1)
````

# Extension 1: Punctuation Should Stay In Place 

__Q__: How should the function perform on consecutive middle white spaces? <br>
__A__: Compress multiple white spaces in the middle of the input into one.

Example 1:

````python
import string

def reverse_words_keep_punctuation(s):
    stack = [] # For words
    queue = [] # For punctuations
    
    N = len(s) 
    word_start, word_end = 0, 0
    punc_start, punc_end = 0, 0 
    i = 0
    while i < N:
        if s[i] == ' ':
            if i > 0:
                if s[i - 1] in string.ascii_letters:
                    stack += [s[word_start:i]]
                elif s[i - 1] in string.punctuation:
                    queue += [(s[punc_start:i], len(stack))]
            i += 1
        elif s[i] in string.ascii_letters:
            if i > 0 and s[i - 1] not in string.ascii_letters:
                word_start = i
                if s[i - 1] in string.punctuation:
                        queue += [(s[punc_start:i], len(stack))]
            i += 1
        elif s[i] in string.punctuation:
            if i > 0: 
                if s[i - 1] not in string.punctuation:
                    punc_start = i
                    if s[i - 1] in string.ascii_letters:
                        stack += [s[word_start:i]]
            i += 1
        else:
            raise Exception('Invalid input')
        
    ans = ""
    counter = 0
    while queue:
        punctuation, word_count = queue.pop(0)
        while counter < word_count:
            word = stack.pop()
            ans += word
            ans += ' '
            counter += 1
            
        if ans and ans [-1] == ' ':
            ans = ans[:(len(ans) - 1)]
        ans += punctuation + ' '

    return ans


print(reverse_words_keep_punctuation("  hello.    word!?  "))  
print(reverse_words_keep_punctuation("  hello.    ,word!?  "))  
print(reverse_words_keep_punctuation("  hello.    word  !?  "))
print(reverse_words_keep_punctuation("  hello  ...    word  !  ?  ")) 
print(reverse_words_keep_punctuation(" ;,  hello  ...    word  !  ?  ")) 
print(reverse_words_keep_punctuation(" ;,  hello  ...    word  !  ?  ...")) 
````
